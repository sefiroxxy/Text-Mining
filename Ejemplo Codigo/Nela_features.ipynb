```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
from nela_features.nela_features import NELAFeatureExtractor
```

```python
#texto de ejemplo de https://pypi.org/project/nela-features/
newsarticle = "Breaking News: Ireland Expected To Become World's First Country To Divest From Fossil Fuels ..." 

nela = NELAFeatureExtractor()

# Extract all feature groups at once
feature_vector, feature_names = nela.extract_all(newsarticle)

# Extract each feature group independently
print("Style:          ", nela.extract_style(newsarticle), "\n")
print("Complexity:     ", nela.extract_complexity(newsarticle), "\n")
print("Bias:           ", nela.extract_bias(newsarticle), "\n")
print("Extract_affect: ", nela.extract_affect(newsarticle), "\n")
print("Extract_moral:  ", nela.extract_moral(newsarticle), "\n")
print("Extract_event:  ", nela.extract_event(newsarticle), "\n")
```
